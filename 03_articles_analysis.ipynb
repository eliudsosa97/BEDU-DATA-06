### Análisis de datos

Saber cuáles de los diferentes sitios web (sources) tienen más interacción

```python
COLUMNS = [
    'source_name',
    'author',
    'engagement_reaction_count',
    'engagement_comment_count',
    'engagement_share_count',
    'engagement_comment_plugin_count'
]

#Limpiar aquellas filas que tienen null en las columnas seleccionadas
data = data[COLUMNS].dropna(axis=0, how='any')
```

```python
grouped_sum = data.groupby('source_name')['engagement_reaction_count'].sum()
grouped_sum.to_frame()
```

```python
grouped_sum = data.groupby('source_name')['engagement_share_count'].sum()
grouped_sum.to_frame()
```

```python
# extraer nueva serie
highest_source = data['source_name'].value_counts()
highest_source.to_frame()
```

```python
labels = list(highest_source.index.values)
print(labels)
```

```python
# Crear gráfica de pastel
plt.pie(highest_source, labels=labels, autopct='%1.1f%%')
plt.title('Fuentes de información')
plt.show()
```
